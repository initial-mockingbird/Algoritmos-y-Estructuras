```python
# We only need to find the GCD of two numbers
def gcd(a, b):
    while b > 0:
        a, b = b, a % b
    return a

def findGCDArray(a): 
    res = a[0]
    for i in a[1:]:
        res = gcd(res, i)
    return res

def findGCDArrayPair(a):
    # Two pointers
    pointA = 0
    pointB = 1

    # New array
    res = []

    # range in the array length
    for _ in range(len(a) - 2):
        if pointB >= len(a): 
            pass
        # find the GCD of the two numbers 
        gcdRes = gcd(a[pointA], a[pointB])

        # Res append to new array 
        res.append(gcdRes)
        
        # Increment on 2 each pointer 
        pointA += 2
        pointB += 2

    return res


print(findGCDArrayPair([2, 20, 40, 80, 90, 9]))
# [2, 10, 22, 40, 80, 22]
#  ^
#     ^

# [2, 10, 22, 40, 80, 20]
#          ^
#               ^

# [2, 10, 22, 40, 80, 40]
#                  ^
#                      ^
print(findGCDArray([4, 10, 12, 22]))
```
