# Takusu Strings

A takusu string is:

1. Composed of only 1s and 0s
2. With the same number of 1s as 0s
3. With no more than 2 consecutive 1s or 0s

Generate all the takusu strings of length `n`:

```
Input: n = 3
Output: ["001","010","011","100","101","110"]
```

````python

TString = Tuple[str,int]

# Meto todos los numeros binarios de longitud n en un arreglo?
# [00, 11, 01, 10]
# [   100,         011, 101, 001, 010, 110]
# [0100,1100,     1011,0011,...]
# imaginandonos que tenmos takuzuStrins_(n-1), anadir 0 o 1 para conseguir akuzuStrins_(n)

# total = 6
# 1
# 01101 1

# Induccion? 
# P(0) == True
# P(n-1) => P(n)
def takusuStrings_(n: int) -> List[str]:
    if n == 0:
        return []
    if n == 1:
        return ['0','1']
    if n == 2:
        return ['00','01','10','11']

    res : List[str] = [] 
    # bottom up approach
    for elem in takusuStrings_(n-1):
        if elem[:2] == '00':
            res.append('1'+elem)
        elif elem[:2] == '01':
            res.append('1'+elem)
            res.append('0'+elem)
        elif elem[:2] == '10':
            res.append('1'+elem)
            res.append('0'+elem)
        elif elem[:2] == '11':
            res.append('0'+elem)
    



    return res



def takusuStrings(n : int) -> List[str]:
    if n == 0:
        return []
    
    tStrings : List[TString] = [("0",-1),("1",1)]
    for i in range(n-1):
        aux : List[TString] = []
        max_diff = n-i
        for (tString,count) in tStrings:
            if abs(count) > max_diff:
                continue
            if tString[:2] != "00":
                aux.append(("0"+tString,count-1))
            if tString[:2] != "11":
                aux.append(("1"+tString,count+1))
        tStrings = [elem for elem in aux]
    
    return [string for (string,count) in tStrings if count == 0]

print(takusuStrings_(6))
````
