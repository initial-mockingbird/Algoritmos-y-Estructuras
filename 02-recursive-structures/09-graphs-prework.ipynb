## Tipos de grafos 
#### Adjencency Matrix

El axis (y) representa el vertice y el axis (x) son los vertices con los que esta conectado 

<img src="./img/pre_work_adjencency.png" width="450">


#### Adjencency List
Es un grafo representado con un diccionario o objeto. Cada propiedad es un vertice y sus "edges" son los que estan dentro del arreglo. Ejemplo:

```json
{
    "a": ["b", "d"],
    "b": ["a", "c"],
    "c": ["b", "e"],
    "e": ["c", "d"],
    "d": ["a", "e"],
}
```

## TODO: BigO de grafos 
Space Complexity

- Con un Adjency Matrix: Cada vertice tiene que guardar todos los vertices que esta conectado, aun cuando no los este usando. ` O(v^2) `

- Con un Adjency List: Aqui solo tenemos que guardar sus vertices ` O(v + e)` => `vertice + la cantidad de vertices` 


Time complexity:
- Adjency Matrix: Si tenemos que agregar un record nuevo, tendriamos que sobre escribir todo la matriz. `O(n)`
- Adjency List: Solo es una propiedad nueva. `O(1)`
