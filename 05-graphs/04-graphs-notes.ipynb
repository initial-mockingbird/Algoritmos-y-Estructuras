```python
# Remove edges
class Graph:
    def __init__(self):
        self.adj_list = {}

    def print_graph(self):
        print('-----------------')
        for vertex in self.adj_list:
            print(vertex, '->', self.adj_list[vertex])
        print('-----------------')

    def add_vertex(self, vertex):
        if vertex != self.adj_list.keys(): 
            self.adj_list[vertex] = []
            return True
        return False

    def add_edge(self, vertex1, vertex2):
        keys = self.adj_list.keys()
        if vertex1 in keys and vertex2 in keys:
            self.adj_list[vertex1].append(vertex2)
            self.adj_list[vertex2].append(vertex1)
            return True
        return False

    # Remover vinculo 
    def remove_edge(self, vertex1, vertex2):
        keys = self.adj_list.keys()
        if vertex1 in keys and vertex2 in keys:
            try:
                self.adj_list[vertex1].remove(vertex2)
                self.adj_list[vertex2].remove(vertex1)
            except ValueError:
                pass
            return True
        return False

testGraph = Graph()
testGraph.add_vertex('A')
testGraph.add_vertex('B')
testGraph.add_vertex('C')
testGraph.add_edge('A', 'B')
testGraph.add_edge('C', 'B')
testGraph.print_graph()
testGraph.remove_edge('A', 'B')
testGraph.print_graph()
```

```python
# Remove vertex 
class Graph:
    def __init__(self):
        self.adj_list = {}

    def print_graph(self):
        for vertex in self.adj_list: 
            print(vertex, "->", self.adj_list[vertex])

    def add_vertex(self, vertex):
        if self.adj_list.keys() != vertex:
            self.adj_list[vertex] = []
            return True
        return False
    
    def add_edge(self, vertex1, vertex2):
        keys = self.adj_list.keys()
        if vertex1 in keys and vertex2 in keys:
            self.adj_list[vertex1].append(vertex2)
            self.adj_list[vertex2].append(vertex1)
            return True
        return False
    
    def remove_vertex(self, vertex):
        edges = self.adj_list[vertex]
        for connected_vertex in edges:
            self.adj_list[connected_vertex].remove(vertex)
        del self.adj_list[vertex]

test_graph = Graph()
test_graph.add_vertex('A')
test_graph.add_vertex('B')
test_graph.add_vertex('C')
test_graph.add_edge('A', 'B')
test_graph.add_edge('A', 'C')
test_graph.add_edge('B', 'C')
test_graph.print_graph()
print('--------')
test_graph.remove_vertex('A')
test_graph.print_graph()
```

# BFS 

- Time complexity O(V + E) = O(Vertix + Edge)
                
- Memory complexity: No esta tan bueno porque estamos haciendo un "Queue" guardando todas las referencias


```python
# Setup basico
class Node:
    def __init__(self, name):
        # Vertice
        self.name = name
        # Edges
        self.adjacency_list = []
        # Fue visitado
        self.visited = False

def breadth_first_search(start_node):
    # FIFO: Primer item que insertamos, sera el primer item que saquemos
    queue = [start_node]

    # Seguimos moviendo hasta que el queue este vacio 
    while queue:
        current_node = queue.pop(0)
        current_node.visited = True

        print(current_node.name)
        print(current_node.visited)

        # Los vecinos del nodo
        for node in current_node.adjacency_list:
            if node.visited == False: 
                queue.append(node)

node_1 = Node("London")
node_2 = Node("New York")
node_3 = Node("Caracas")
node_4 = Node("San Francisco")
node_5 = Node("Madrid")

node_1.adjacency_list = [node_2]
node_2.adjacency_list = [node_1, node_3]
node_3.adjacency_list = [node_2, node_4]
node_4.adjacency_list = [node_5, node_3]
node_5.adjacency_list = [node_4]

breadth_first_search(node_1)
```

# DFS
DFS explora todas las posibilidades de una linea antes de hacer un backtracking. A diferencia del BFS donde va capa por capa 

Complejidad de tiempo: O(Vertices + Edges)
Complejidad en memoria: un poquito mejor que el BFS 


```python
# Base 
class Node:
    def __init__(self, name): 
        self.name = name
        self.adjacency_list = []
        self.visited = False

node_1 = Node('A')
node_2 = Node('B')
node_3 = Node('C')
node_4 = Node('D')
node_5 = Node('E')

node_1.adjacency_list.append(node_2)
node_2.adjacency_list.append(node_3)
node_3.adjacency_list.append(node_4)
node_4.adjacency_list.append(node_5)



# DFS Recursivo 
def depth_first_search_recursive(vertex):
    print(vertex.name)
    vertex.visited = True
    for edges in vertex.adjacency_list: 
        depth_first_search_recursive(edges)

depth_first_search_recursive(node_1)

########################################################33


# Reset
node_1.visited = False
node_2.visited = False
node_3.visited = False
node_4.visited = False
node_5.visited = False

########################################################33



# DFS Iterando
def depth_first_search_iterative(start_node): 
    
    # LIFO = Ultimo item insertado, es el primero en abandonar
    stack = [start_node]

    while stack:
        node = stack.pop()
        node.visited = True
        print(node.name)
        for edge_node in node.adjacency_list:
            if edge_node.visited == False:
                stack.append(edge_node)


print('------------ DFS with Stacks')
depth_first_search_iterative(node_1)



#########################



```
